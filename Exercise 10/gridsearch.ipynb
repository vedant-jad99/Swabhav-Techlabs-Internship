```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
```

```python
df=pd.read_csv('final_cars.csv')
Y=df.price
X=df.drop('price',axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
hyperparameters={'max_depth':[2,4,6,8],'min_samples_leaf':[5,4,8,10,15]}
```

```python
gridseach=GridSearchCV(DecisionTreeRegressor(),hyperparameters,cv=5)
```

```python
X=pd.get_dummies(X)
X
X_train , X_test , Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
gridseach.fit(X_train,Y_train)
```

```python
gridseach.best_params_
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import r2_score
```

```python
from sklearn.preprocessing import  scale
```

```python
X
```

```python
X
```

```python
X['engine-size']=scale(X['engine-size'])
X['horsepower']=scale(X['horsepower'])
X['city-mpg']=scale(X['city-mpg'])
```

```python
X
```

```python
X_train , X_test , Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
model=SGDRegressor(random_state=100)
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
r2_score(Y_test,model.predict(X_test))
```

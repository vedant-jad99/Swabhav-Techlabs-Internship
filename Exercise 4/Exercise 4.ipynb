## Task 1
Create an array of 5 * 5 with random numbers in the range 1 to 100 and display how many numbers in the array are even numbers.

```python
import numpy as np

arr = np.random.randint(1, 100, (5, 5))
even = (arr%2 == 0)
print(even, '\n\n', arr)
print("\nThe number of even numbers :", np.count_nonzero(even))
```

## Task 2

1. Make a 2D array of random integers count how many values are less than 6

2. Make a 2D array of random integers count how many values are less than 6 in each row

3. From 2D array check if there are any values greater than 8? (output: True or false)

4. From 2D array check if there are any values less than 0? (output: True or false)

5. From 2D array check all values less than 10? (output: True or false)

6. Are all values in each row less than 8? (output:True/False)

```python
rand_arr = np.random.randint(-5, 8, (5, 5))
val_lt_6 = (rand_arr < 6)
print(rand_arr, "\n\n", val_lt_6)
print("\nNumber of values less than 6 :", np.count_nonzero(val_lt_6), "\n\n")

for i, row in enumerate(val_lt_6):
    print("Row :", i, " -", np.count_nonzero(row))
    
print("\n\nAre any values greater than 8 :", np.count_nonzero(rand_arr > 8) > 0)
print("\n\nAre any values less than 0 :", np.count_nonzero(rand_arr < 0) > 0)
print("\n\nAre all values less than 10 :", np.count_nonzero(rand_arr < 10) == 25)
print("\n\nAre all values in each row less than 8 :", np.count_nonzero(np.array([np.count_nonzero(i < 8) for i in rand_arr]) == 5) == 5)
```

## Task 4

Take an array of 10 rows with 5 columns (fill it with random numbers) and display average for each row and which row is having highest average (if 2 students got the same highest average, display both the student rows)

```python
arr = np.random.randint(-10, 10, (10, 5))
print("Arr: \n", arr)
avg_arr = np.average(arr, axis=1)
print("\nAverages of each row:\n", avg_arr)
print("\nHighest average", arr[avg_arr == max(avg_arr)])
```

## Task 6

- Take an image, split it into two halves vertically and interchange those two parts 

- Load an image and interchange red with green.

```python
from matplotlib import pyplot as plt, image as im
import os

image = None
for i in os.listdir(os.getcwd()):
    if ".jpg" in i:
        image = im.imread(os.path.join(os.getcwd(), i))
        
plt.imshow(image)

new_arr = np.vsplit(image, 2)
```

```python
new_arr[0], new_arr[1] = new_arr[1], new_arr[0]
plt.imshow(np.vstack((new_arr[0], new_arr[1])))
```

```python
img = image.copy()
img[:,:,0], img[:,:,1] = img[:,:,1], img[:,:,0]
plt.imshow(img)
```
